```python
import matplotlib.pyplot as plt
```

```python
grouped_data = data.groupby('City')['Total Number of Accident'].mean()
plt.bar(grouped_data.index, grouped_data.values,width=0.8)

plt.xlabel('City',fontsize=12)
plt.ylabel('Total Number of Accident',fontsize=12)
plt.title('Histogram of Data',fontsize=12)
plt.xticks(rotation=90, ha='center')

plt.show()
```

```python
grouped_data = data.groupby('Month - Year')['Total Number of Accident'].mean()
plt.bar(grouped_data.index, grouped_data.values,width=0.8)

plt.xlabel('Month - Year',fontsize=12)
plt.ylabel('Total Number of Accident',fontsize=12)
plt.title('Histogram of Data',fontsize=12)
plt.xticks(rotation=90, ha='center')

plt.show()
```

```python
grouped_data = data.groupby('Month - Year')['Accident Fatal'].mean()
plt.bar(grouped_data.index, grouped_data.values,width=0.8)

plt.xlabel('Month - Year',fontsize=12)
plt.ylabel('Accident Fatal',fontsize=12)
plt.title('Histogram of Data',fontsize=12)
plt.xticks(rotation=90, ha='center')

plt.show()
```

```python
grouped_data = data.groupby('City')[['Total Number of Accident', 'Accident Fatal', 'Accident Non-Fatal', 'Person Killed', 'Person Injured', 'Total Number of vehicles Involves']].mean()

plt.bar(grouped_data.index, grouped_data['Total Number of Accident'], width=0.8)

plt.xlabel('City', fontsize=12)
plt.ylabel('Total Number of Accident', fontsize=12)
plt.title('Histogram of Data', fontsize=12)
plt.xticks(rotation=90, ha='center')

plt.show()
```
